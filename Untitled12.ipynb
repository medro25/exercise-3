```python
fd_1 = indexed.selectExpr(
    "cast(dateTime as string) as dateTime",
    "cast(relative_humidity as float) as relative_humidity",
    "cast(wind_speed as float) as wind_speed",
    "cast(indexedTime as double) as indexedTime",
    "cast(air_temperature as float) as label",
    "cast(msl as float) as msl"
)
```

```python

#Q: Find alternative approach to do the same job
from pyspark.sql.functions import expr

# Create a new DataFrame with the vectorized features
trSet = fd_1.selectExpr(
    "*",
    "array(relative_humidity, wind_speed, indexedTime, msl) as features"
).select("features", "label")

print(trSet)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col

# Step 1: Remove rows with null values
fd_1_filtered = fd_1.dropna(subset=['relative_humidity', 'wind_speed', 'indexedTime', 'msl'])

# Step 2: Create a VectorAssembler to combine input columns into a vector column
vectorAssembler = VectorAssembler(
    inputCols=['relative_humidity', 'wind_speed', 'indexedTime', 'msl'],
    outputCol='features',
    handleInvalid='skip'  # or 'skip'
)

# Step 3: Apply the VectorAssembler transformation to the filtered dataset
vhouse_df = vectorAssembler.transform(fd_1_filtered)

# Step 4: Select 'features' and 'label' columns for the training set
trSet = vhouse_df.select("features", "label")

# Step 5: Split the data into training and test sets (30% held out for testing)
(trainingData, testData) = trSet.randomSplit([0.7, 0.3])

# Step 6: Create a GBTRegressor model
gbt = GBTRegressor(featuresCol="features", maxIter=30, maxDepth=11)

# Step 7: Train the GBTRegressor model
model = gbt.fit(trainingData)

# Step 8: Make predictions on the test set
predictions = model.transform(testData)

# Step 9: Evaluate the model
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE):", rmse)

```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit

# Step 1: Remove rows with null values
fd_1_filtered = fd_1.dropna(subset=['relative_humidity', 'wind_speed', 'indexedTime', 'msl'])

# Step 2: Create a VectorAssembler to combine input columns into a vector column
vectorAssembler = VectorAssembler(
    inputCols=['relative_humidity', 'wind_speed', 'indexedTime', 'msl'],
    outputCol='features',
    handleInvalid='skip'
)

# Step 3: Apply the VectorAssembler transformation to the filtered dataset
vhouse_df = vectorAssembler.transform(fd_1_filtered)

# Step 4: Select 'features' and 'label' columns for the training set
trSet = vhouse_df.select("features", "label")

# Step 5: Create a GBTRegressor model
gbt = GBTRegressor(featuresCol="features", maxIter=30, maxDepth=11)

# Step 6: Create a RegressionEvaluator
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")

# Step 7: Define the parameter grid for tuning
paramGrid = ParamGridBuilder() \
    .addGrid(gbt.maxIter, [10, 20, 30]) \
    .addGrid(gbt.maxDepth, [5, 10, 15]) \
    .build()

# Step 8: Create a TrainValidationSplit
tvs = TrainValidationSplit(estimator=gbt,
                           estimatorParamMaps=paramGrid,
                           evaluator=evaluator,
                           trainRatio=0.8)  # Set the ratio for train/validation split

# Step 9: Fit the TrainValidationSplit to the training data
tvModel = tvs.fit(trSet)

# Step 10: Get the best model from train/validation
bestModel = tvModel.bestModel

# Step 11: Make predictions on the test data using the best model
predictions = bestModel.transform(trSet)

# Step 12: Evaluate the model performance
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
"""Q: Try to perform weather prediction using DateTime separately as Day(Mon, Tue etc.),
Time (in hours , 11:00, 12:00 etc.)

"""

from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col, dayofweek, hour
from pyspark.sql import SparkSession
from pyspark.sql.functions import dayofmonth, hour, to_timestamp
from pyspark.sql.functions import date_format

# Remove the null values
fd_1_filtered = fd_1_filtered.dropna(subset=['relative_humidity', 'wind_speed', 'indexedTime', 'msl', 'DateTime'])

# Convert DateTime to timestamp and extract day, hour, and day_name
fd_1_filtered = fd_1_filtered.withColumn('DateTime', to_timestamp('DateTime', 'dd-MMM-yyyy HH:mm'))
fd_1_filtered = fd_1_filtered.withColumn('day', dayofmonth('DateTime'))
fd_1_filtered = fd_1_filtered.withColumn('hour', hour('DateTime'))
fd_1_filtered = fd_1_filtered.withColumn('day_name', date_format('DateTime', 'EEEE'))

# Show the updated dataframe
fd_1_filtered.show()

# Create a VectorAssembler to combine input columns into a vector column
vectorAssembler = VectorAssembler(
    inputCols=['relative_humidity', 'wind_speed', 'day', 'hour', 'msl'],
    outputCol='features',
    handleInvalid='skip'
)

# Apply the VectorAssembler transformation to the filtered dataset
vhouse_df = vectorAssembler.transform(fd_1_filtered)

# Select 'features', 'hour', 'day_name', and 'label' columns for the training set
trSet = vhouse_df.select("features", "hour", "day_name", "label")
trSet.show()

# Split the data into training and test sets (70% training, 30% testing)
(trainingData, testData) = trSet.randomSplit([0.7, 0.3])

# Create a GBTRegressor model
gbt = GBTRegressor(featuresCol="features", maxIter=30, maxDepth=11, maxBins=32, labelCol="label")

# Train the GBTRegressor model
model = gbt.fit(trainingData)

# Make predictions on the test data and select the predicted 'hour', 'day_name', and 'prediction' columns
predictions = model.transform(testData)
predictedHoursAndDays = predictions.select("hour", "day_name", "prediction")
predictedHoursAndDays.show()

# Evaluate the model performance
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)

# Print the root mean squared error (RMSE)
print("Root Mean Squared Error (RMSE):", rmse)

```
